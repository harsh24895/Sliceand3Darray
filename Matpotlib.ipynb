# Matplotlib

```python
import matplotlib.pyplot as plt

%matplotlib inline
x = list(range(0,10))
y = list(range(-10,0))     

plt.plot(x,y)
```

```python
import matplotlib.pyplot as plt

%matplotlib inline
a = [0,-100,25,67,-32]
b = [0,3,5,8,9]
plt.plot(a,b)
```

```python
#this will zoom the upper plot graph
# ex:x-axis 0,75 and y-axis: 2,8
plt.axis([0, 75, 2, 8])
plt.plot(a,b)
```

```python
#title the plot and give the label 
plt.title("line graph")
plt.xlabel("Array A")
plt.ylabel("Array B")
plt.plot(a,b)
```

```python
#change the label of x- axis 
plt.title("line graph")
plt.xlabel("Array A")
plt.ylabel("Array B")
plt.xticks((-100, -75, -50, -25, 0, 25, 50, 75),('h','e','g','k','j','h','g','k'))
plt.plot(a,b)
```

```python
#to set the color for the graph
plt.plot(a,b, color="red")
```
